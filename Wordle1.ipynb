```python
import random
import string 
    
# Create wordle-word bank and returns random word
def Game_Word():
    # Storing the sets of punctuation in variable result 
    result = string.punctuation
    
    # Create Word Bank of 5-letter Words
    with open ('All_Words.txt') as contents:
        base = contents.readlines()
        
        f = open('Wordle_Bank.txt','w')

        for i in base:
            check = True
            i = i.replace('\n','')
            
            # Checks if any punctuation in word
            for j in i:
                if j in result:
                    check=False
                    
            # Adds word to Wordle Bank
            if len(i) == 5 and check:
                f.write('{}\n'.format(i.lower()))

        f.close()

    # Obtain random 5-letter word
    with open ('Wordle_Bank.txt') as contents:
        wbank = contents.readlines()
        word = (random.choice(wbank)).replace('\n','')
        
    return wbank,word
```

```python
# Outputs guess color-coded
# Green: letter in right place
# Yellow: letter in word but in wrong place
# Gray: Wrong letter, wrong place

def Output_Color(guess,word,alphabet): 
    # Check guess with word   
    output = ""
    # Counts number of correct letters
    count = []
    gletter =[]
    yletter =[]
    # alphabet = alphabet.replace("\033[0;30;42m ","")
    # alphabet = alphabet.replace("\033[0;30;43m ","")
    
    for i in range(len(word)):
        # Checks if letter is in word and in right place
        if word[i] == guess[i]:
            output+=("\033[0;30;42m {}\033[0;0m".format(guess[i]))
            gvalue = "\033[0;30;42m{}\033[0;0m".format(guess[i])
            yvalue = "\033[0;30;43m{}\033[0;0m".format(guess[i])
            if yvalue in alphabet:
                alphabet=alphabet.replace(yvalue,gvalue)
            else:
                alphabet=alphabet.replace(" "+ guess[i]," "+ gvalue)
            # print(alphabet)
            count.append(i) 
            gletter.append(guess[i])
            
        # Checks if letter is in word
        elif guess[i] in word:
            output+=("\033[0;30;43m {}\033[0;0m".format(guess[i]))
            gvalue = "\033[0;30;42m{}\033[0;0m".format(guess[i])
            yvalue = "\033[0;30;43m{}\033[0;0m".format(guess[i])
            if gvalue in alphabet:
                alphabet=alphabet.replace(gvalue,yvalue)
            else:
                alphabet=alphabet.replace(" "+ guess[i]," "+ yvalue)
            # alphabet=alphabet.replace(guess[i],yvalue)
            yletter+=guess[i]
        
        # Output letter is not in word.
        else:
            output+=("\033[0;30;47m {}\033[0;0m".format(guess[i]))
            grayvalue = "\033[0;30;47m{}\033[0;0m".format(guess[i])
            if grayvalue not in alphabet:
                alphabet=alphabet.replace(" "+ guess[i]," "+ grayvalue)
            # alphabet=alphabet.replace(guess[i],grayvalue)
    return(output,count,gletter,yletter,alphabet)
```

```python
# Retrieves user input for guess
def Guess_Attempt():
    guess = input("Enter your guess: ")

    # Checks if word is 5-letters and in word bank
    check = guess+'\n' in wbank       
    
    while len(guess)!= 5 or check == False:
        if len(guess)!= 5:
            guess = input("Please enter a 5-letter word: ")
        elif check == False:
            guess = input("That is not a word. Please enter your guess: ")
            check = guess+'\n' in wbank
          
    return(guess)
```

```python
# Retrieves user input that includes previous yellow and green letters.
def Guess_Hard_Mode(gletter,yletter,count):
    guess = input("Enter your guess: ").lower()

    # Checks if word is 5-letters, in word bank, follows green and yellow letters
    check = guess+'\n' in wbank       
    
    gcheck = len(gletter)>0
    ycheck = len(yletter)>0
        
    while len(guess)!= 5 or check == False or gcheck or ycheck:
        # Check for 5-letter word
        if len(guess)!= 5:
            guess = input("Please enter a 5-letter word: ").lower()
            gcheck = len(gletter)>0
            ycheck = len(yletter)>0            
        
        # Check if input is a word
        elif check == False:
            guess = input("That is not a word.\nPlease enter your guess: ").lower()
            check = guess+'\n' in wbank
            gcheck = len(gletter)>0
            ycheck = len(yletter)>0    
        
        # Check if green letters are kept in right place
        elif gcheck:
            # print("Count of green letter: {}".format(count))
            for i in range(len(count)):
                # print("Letter is {}".format(guess[count[i]]))
                
                if guess[count[i]] != gletter[i]:
                    guess = input("Green letter(s) are not in correct place.\nPlease enter your guess: ").lower()
                    break
                else:
                    gcheck = False

        # Check if yellow letter are being used
        elif ycheck:
            for i in range(len(yletter)):
                if yletter[i] not in guess:
                    guess = input("Yellow letter(s) are not included in guess.\nPlease enter your guess: ").lower()
                    break
                else:
                    ycheck = False
        
    return(guess)
```
